### Transform Membership Data
1. Extract customer_id from the file path
2. Write transformed data to the silver schema

```sql
select * from gizmobox.bronze.v_membership
```

#####1. Extract customer_id from the file path using regexp_extract()

```sql
select regexp_extract(path, '.*/([0-9]+)\\.png',1) as customer_id,
       date_format(modificationTime, 'yyyy-MM-dd') as modified_date,
       date_format(modificationTime, 'HH:mm:ss') as modified_time,
       content as membership_card
from gizmobox.bronze.v_membership
```

#####2. Write transformed data to the silver schema

```sql
create table gizmobox.silver.memberships
as
select regexp_extract(path, '.*/([0-9]+)\\.png',1) as customer_id,
       date_format(modificationTime, 'yyyy-MM-dd') as modified_date,
       date_format(modificationTime, 'HH:mm:ss') as modified_time,
       content as membership_card
from gizmobox.bronze.v_membership
```

```sql
select *  from gizmobox.silver.memberships
```
